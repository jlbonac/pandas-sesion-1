```python
# quiero una funcion que me permita categorizar la edad de los pasajeros, ninos menores de 11, adolecentes de 12 a 17, adultos de 18 a 59 y adultos mayores de 60 dentro de age-cat y muestrame el resultado 
def categorize_age(age):
    if age < 11:
        return "Niño"
    elif age >= 12 and age < 18:
        return "Adolescente"
    elif age >= 18 and age <= 59:
        return "Adulto"
    elif age >= 60:
        return "Adulto mayor"

df_titanic3["Age-cat"] = df_titanic3["Age"].apply(categorize_age)
df_titanic3["Age-cat"].value_counts()
```

```python
# completar los datos nulos de la columna Embarked con la moda
df_titanic4["Embarked"].fillna(df_titanic4["Embarked"].mode()[0], inplace=True)
```

```python
# completar los datos nulos de la columna Age con la media
df_titanic4["Age"].fillna(df_titanic4["Age"].mean(), inplace=True)
```
